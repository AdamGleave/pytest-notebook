# Markdown Cell

```python
# code cell, no output
import sys
from IPython import display
```

```python
# code cell + stdout
print('hallo')
```

```python
# code cell + stdout + stdout
print('hallo1')
print('hallo2')
```

```python
# code cell + stderr
print('hallo', file=sys.stderr)
```

```python
# code cell + stderr + stdout
print('hallo1', file=sys.stderr)
print('hallo2')
```

```python
# code cell + latex
display.Latex('\\textit{hallo}')
```

```python
# code cell + html
display.HTML("""
<div class="section" id="submodules">
    <h2>Submodules<a class="headerlink" href="#submodules" title="Permalink to this headline">¶</a></h2>
</div>""")
```

```python
# code cell + json
display.JSON({'a': [1, 2, 3, 4,], 'b': {'inner1': 'helloworld', 'inner2': 'foobar'}})
```

```python
# code cell + markdown
display.Markdown('## Header\n\n- bullet')
```

```python
# code cell + SVG
display.SVG("""
<svg height="100" width="100">
  <circle cx="50" cy="50" r="40" stroke="black" stroke-width="3" fill="red" />
</svg> """)
```

```python
# code cell + Math
display.Math("a = b + c")
```

```python
# code cell + png
display.Image('https://octodex.github.com/images/yaktocat.png', width=300, unconfined=True)
```

```python
# code cell + jpg
display.Image('https://i.ytimg.com/vi/j22DmsZEv30/maxresdefault.jpg', width=500, unconfined=True)
```

```python
display.display(display.Image('https://octodex.github.com/images/yaktocat.png', width=300, unconfined=False))
print('hallo1', file=sys.stderr)
print('hallo2')
display.Image('https://i.ytimg.com/vi/j22DmsZEv30/maxresdefault.jpg', width=500, unconfined=True)
```
